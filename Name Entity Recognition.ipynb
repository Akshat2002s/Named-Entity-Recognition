```python
import spacy 
from spacy import displacy
import numpy as np
import pandas as pd
```

```python
nlp= spacy.load("en_core_web_md")
```

```python
with open("text.txt", "r") as f:
    document=f.readlines()
```

```python
type(document)
```

```python
text=" "
text=text.join(document)
```

```python
text
```

```python
doc=nlp(text)

entities=[]
labels=[]

for ent in doc.ents:
    entities.append(ent)
    labels.append(ent.label_)

df=pd.DataFrame({'Entities':entities,'Labels':labels})
```

```python
df.head()
```

```python
df.to_csv('entity.csv', index=False)
```

```python
import nltk
```

```python
words = nltk.word_tokenize(text)
words
```

```python
pos_tags = nltk.pos_tag(words)
pos_tags
```

```python
nltk.help.upenn_tagset('VBZ')
```

```python
chunks = nltk.ne_chunk(pos_tags, binary=True)
for chunk in chunks:
    print(chunk)
```

```python
entities =[]
labels =[]
for chunk in chunks:
    if hasattr(chunk,'label'):
        #print(chunk)
        entities.append(' '.join(c[0] for c in chunk))
        labels.append(chunk.label())
        
entities_labels = list(set(zip(entities, labels)))
entities_df = pd.DataFrame(entities_labels)
entities_df.columns = ["Entities","Labels"]
entities_df
```

```python
entities_df.to_csv('binary_entity.csv', index=False)
```
